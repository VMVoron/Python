Формулу Стерджесса для равных интервалов

```python
import random
```

Случайная группа из 1000 чисел от 0 до 999

```python
n = 1000
```

```python
group = [int(random.randrange(1000)) for i in range(n)]
```

```python
import math
```

```python
k = 0
```

```python
k = 1 + 3.322 * math.log10(n)
```

```python
k
```

Или, примерно:

```python
k = 1 + 1.4 * math.log(n)
```

```python
k
```

Примем k = 7.644

Величина (длина интервала) дельта

```python
d = (max(group) - min(group)) // k
```

```python
d
```

```python
d*k
```

```python
d*math.ceil(k)
```

```python
group.sort()
```

```python
g = [int(i) for i in range(int(math.ceil(k)))]
```

```python
len(g)
```

```python
a = 0
b = int(d)
for i in range(0, len(g)+1):
    print(a, 'a', b, 'b')
    g[i] = group[a:b]
    a = b
    b = int(d)*(i+2)
    if b > 1000:
        b = 1000
    if a == 1000 and b == 1000:
        break
```

```python
for i in range(len(g)):
    print('Набор', i, ':', g[i], end='\n')
```
